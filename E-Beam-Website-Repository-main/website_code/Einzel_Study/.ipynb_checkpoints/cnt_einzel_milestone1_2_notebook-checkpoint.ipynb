
# Research-oriented axisymmetric electrostatics notebook for a CNT-driven einzel-lens geometry

This notebook upgrades the earlier prototype in four ways:

1. **Explicit dielectric physics** through the variable-permittivity axisymmetric equation  
   $$
   \frac{1}{r}\frac{\partial}{\partial r}\!\left(r\,\epsilon(r,z)\,\frac{\partial V}{\partial r}\right)
   +
   \frac{\partial}{\partial z}\!\left(\epsilon(r,z)\,\frac{\partial V}{\partial z}\right)=0 .
   $$

2. **More realistic geometry**
   - finite-thickness ring electrodes,
   - rounded electrode corners,
   - an explicit SiO$_2$ dielectric pedestal,
   - and a protruding effective CNT emitter (cylindrical stem + hemispherical cap).

3. **Defensible electrostatics diagnostics**
   - domain-truncation / boundary-sensitivity study,
   - mesh-refinement study,
   - convergence checks on lens-scale observables.

4. **Emission coupling**
   - local electric field extraction near the CNT surface,
   - Fowler–Nordheim current-density evaluation,
   - and axisymmetric surface-integration of the emitted current.

---

## Important modeling note

A truly single CNT tip involves a **nanometer-scale radius** embedded in a **millimeter-scale lens**. A uniform structured grid cannot resolve that multiscale range efficiently. This notebook therefore uses:

- an **effective protruding CNT emitter geometry** that is resolvable on the lens-scale mesh, and
- an optional **sub-grid enhancement factor** $\beta_{\mathrm{subgrid}}$ in the Fowler–Nordheim model to represent unresolved nanoscale enhancement.

That makes this notebook a strong **pre-trajectory electrostatics and emission workflow**, but not yet a full nanoscale emitter simulation.

---

## What is and is not claimed here

This notebook is designed to make the electrostatics portion substantially more defensible. It does **not** yet include:

- space charge,
- self-consistent emission back-reaction,
- particle tracing,
- or experimental / external-solver validation.

Those are natural next steps after this notebook.



## Governing equations

### Axisymmetric electrostatics with explicit dielectric

We solve
$$
\nabla\cdot\left(\epsilon \nabla V\right)=0
$$
in axisymmetric $(r,z)$ coordinates, which becomes
$$
\frac{1}{r}\frac{\partial}{\partial r}\!\left(r\,\epsilon(r,z)\,\frac{\partial V}{\partial r}\right)
+
\frac{\partial}{\partial z}\!\left(\epsilon(r,z)\,\frac{\partial V}{\partial z}\right)=0 .
$$

The electric field is then
$$
\mathbf{E}(r,z)=-\nabla V(r,z),
\qquad
E_r=-\frac{\partial V}{\partial r},
\qquad
E_z=-\frac{\partial V}{\partial z}.
$$

### Fowler–Nordheim coupling

For a local normal field magnitude $F_n$, the Fowler–Nordheim current density is modeled as
$$
J_{\mathrm{FN}}(F_n)
=
\frac{A_{\mathrm{FN}}}{\phi}\,F_n^2
\exp\!\left[-\,B_{\mathrm{FN}}\frac{\phi^{3/2}}{F_n}\right],
$$
where

- $\phi$ is the work function in eV,
- $A_{\mathrm{FN}} = 1.541434\times 10^{-6}\ \mathrm{A\,eV\,V^{-2}}$,
- $B_{\mathrm{FN}} = 6.830890\times 10^9\ \mathrm{V\,m^{-1}\,eV^{-3/2}}$.

Because the lens-scale grid does not resolve a true nm-scale CNT apex, we optionally use
$$
F_{\mathrm{FN}} = \beta_{\mathrm{subgrid}}\,F_n
$$
inside the Fowler–Nordheim law.

### What is used as the "defensible electrostatics" convergence target?

The strongest corner fields converge slowly, even with rounded geometry, because local geometric singularity structure still matters. For that reason, this notebook checks convergence primarily on **lens-scale observables**:

- axis potential $V(0,z)$ in the lens region,
- the integrated axis field $\int E_z(0,z)\,dz$ across the lens region,
- the peak axis potential and its location.

These are the quantities most relevant for beam optics before trajectory tracing.


## Geometry model

The geometry implemented in this notebook is a more realistic axisymmetric surrogate of the sketched setup:

- the cathode plane is at $z=0$ and held at $-V_{\mathrm{CNT}}$,
- the CNT emitter is an **effective protruding conductor**:
  - cylindrical stem,
  - hemispherical cap,
- a SiO$_2$ pedestal surrounds the CNT base,
- the einzel lens is represented by **three finite-thickness annular ring electrodes** with rounded cross-section corners,
- the outer radial boundary and top axial boundary are grounded truncation boundaries placed sufficiently far away to test domain sensitivity.

The result is still a reduced axisymmetric model, but it is substantially closer to a research workflow than the earlier wall-segment approximation.


```python

def rounded_rect_mask(RR, ZZ, r0, r1, z0, z1, rad):
    '''
    Rounded rectangle in the (r,z) cross-section, using a signed-distance-style test.
    '''
    cx = 0.5 * (r0 + r1)
    cz = 0.5 * (z0 + z1)
    hx = 0.5 * (r1 - r0)
    hz = 0.5 * (z1 - z0)

    rad = max(0.0, min(rad, hx, hz))

    qx = np.abs(RR - cx) - (hx - rad)
    qz = np.abs(ZZ - cz) - (hz - rad)

    sdf = np.sqrt(np.maximum(qx, 0.0)**2 + np.maximum(qz, 0.0)**2)         + np.minimum(np.maximum(qx, qz), 0.0) - rad

    return sdf <= 0.0


def build_geometry(r, z, params):
    RR, ZZ = np.meshgrid(r, z, indexing="ij")
    Nr, Nz = RR.shape

    eps_rel = np.ones((Nr, Nz), dtype=float)
    material_id = np.zeros((Nr, Nz), dtype=np.int8)   # 0 vacuum, 1 oxide
    is_dir = np.zeros((Nr, Nz), dtype=bool)
    V_bc = np.zeros((Nr, Nz), dtype=float)
    conductor_id = np.zeros((Nr, Nz), dtype=np.int8)  # 0 none, 1 CNT/cathode, 2 e1, 3 e2, 4 e3

    # Bottom cathode plane
    is_dir[:, 0] = True
    V_bc[:, 0] = -params["V_cnt"]
    conductor_id[:, 0] = 1

    # Outer and top truncation boundaries
    is_dir[-1, :] = True
    V_bc[-1, :] = 0.0

    is_dir[:, -1] = True
    V_bc[:, -1] = 0.0

    # Dielectric pedestal
    oxide_mask = (RR <= params["r_oxide"]) & (ZZ <= params["h_oxide"])
    material_id[oxide_mask] = 1
    eps_rel[oxide_mask] = params["eps_r_oxide"]

    # Effective CNT emitter: cylindrical stem + hemispherical cap
    r_cnt = params["r_cnt"]
    h_cnt_total = params["h_cnt_total"]
    h_cyl = max(h_cnt_total - r_cnt, 0.0)

    cnt_cyl = (RR <= r_cnt) & (ZZ <= h_cyl)
    cnt_cap = (ZZ >= h_cyl) & ((RR**2 + (ZZ - h_cyl)**2) <= r_cnt**2)
    cnt_mask = cnt_cyl | cnt_cap

    is_dir[cnt_mask] = True
    V_bc[cnt_mask] = -params["V_cnt"]
    conductor_id[cnt_mask] = 1

    # Conductors override dielectric labeling
    material_id[cnt_mask] = 0
    eps_rel[cnt_mask] = 1.0

    # Annular ring electrodes
    e1_mask = rounded_rect_mask(
        RR, ZZ,
        params["a_inner"], params["r_e_outer"],
        params["z1_start"], params["z1_end"],
        params["edge_radius"]
    )
    e2_mask = rounded_rect_mask(
        RR, ZZ,
        params["a_inner"], params["r_e_outer"],
        params["z2_start"], params["z2_end"],
        params["edge_radius"]
    )
    e3_mask = rounded_rect_mask(
        RR, ZZ,
        params["a_inner"], params["r_e_outer"],
        params["z3_start"], params["z3_end"],
        params["edge_radius"]
    )

    for mask, cid, val in [
        (e1_mask, 2, 0.0),
        (e2_mask, 3, params["V_mid"]),
        (e3_mask, 4, 0.0),
    ]:
        is_dir[mask] = True
        V_bc[mask] = val
        conductor_id[mask] = cid
        material_id[mask] = 0
        eps_rel[mask] = 1.0

    return {
        "RR": RR,
        "ZZ": ZZ,
        "eps_rel": eps_rel,
        "material_id": material_id,
        "is_dir": is_dir,
        "V_bc": V_bc,
        "conductor_id": conductor_id,
        "cnt_mask": (conductor_id == 1),
    }


def plot_geometry_maps(r, z, geom):
    fig, axs = plt.subplots(1, 3, figsize=(16, 4.8), constrained_layout=True)

    # Material map
    material_plot = np.where(geom["material_id"] == 1, 1.0, 0.0)
    pcm0 = axs[0].pcolormesh(z * 1e3, r * 1e3, material_plot, shading="auto", cmap="Blues")
    axs[0].contour(z * 1e3, r * 1e3, (geom["conductor_id"] > 0).astype(float), levels=[0.5], colors="k", linewidths=0.9)
    axs[0].set_title("Dielectric map (blue = SiO$_2$)")
    axs[0].set_xlabel("z [mm]")
    axs[0].set_ylabel("r [mm]")

    # Boundary / conductor potentials
    bc_plot = np.where(geom["is_dir"], geom["V_bc"], np.nan)
    pcm1 = axs[1].pcolormesh(z * 1e3, r * 1e3, bc_plot, shading="auto", cmap="coolwarm")
    fig.colorbar(pcm1, ax=axs[1], label="Dirichlet value [V]")
    axs[1].set_title("Conductors and truncation boundaries")
    axs[1].set_xlabel("z [mm]")
    axs[1].set_ylabel("r [mm]")

    # Relative permittivity
    pcm2 = axs[2].pcolormesh(z * 1e3, r * 1e3, geom["eps_rel"], shading="auto", cmap="viridis")
    fig.colorbar(pcm2, ax=axs[2], label=r"$\epsilon_r$")
    axs[2].contour(z * 1e3, r * 1e3, (geom["conductor_id"] > 0).astype(float), levels=[0.5], colors="w", linewidths=0.7)
    axs[2].set_title("Relative permittivity map")
    axs[2].set_xlabel("z [mm]")
    axs[2].set_ylabel("r [mm]")

    plt.show()
```


## Discretization

A uniform structured grid is used in $(r,z)$, but the operator includes the **axisymmetric** and **variable-permittivity** structure.

At interior points, the discretization uses face-based harmonic averaging of $\epsilon$:
$$
\epsilon_{i+1/2,j}
=
\frac{2\,\epsilon_{i,j}\,\epsilon_{i+1,j}}{\epsilon_{i,j}+\epsilon_{i+1,j}},
\qquad
\epsilon_{i,j+1/2}
=
\frac{2\,\epsilon_{i,j}\,\epsilon_{i,j+1}}{\epsilon_{i,j}+\epsilon_{i,j+1}}.
$$

This is the standard conservative choice for interface problems because it preserves the correct continuity of normal flux across dielectric boundaries.

At the axis $r=0$, the symmetry condition
$$
\left.\frac{\partial V}{\partial r}\right|_{r=0}=0
$$
is enforced through the special axis stencil.


```python

def harmonic_mean(a, b):
    return 2.0 * a * b / (a + b) if (a + b) != 0.0 else 0.0


def build_axisymmetric_variable_eps_matrix(r, z, eps_rel, is_dir, V_bc):
    '''
    Build A x = b for the variable-permittivity axisymmetric equation

        (1/r) d/dr [ r eps dV/dr ] + d/dz [ eps dV/dz ] = 0

    on a uniform structured grid.
    '''
    Nr, Nz = len(r), len(z)
    dr = r[1] - r[0]
    dz = z[1] - z[0]

    unknown_id = -np.ones((Nr, Nz), dtype=int)
    coords = []
    k = 0

    for i in range(Nr):
        for j in range(Nz):
            if not is_dir[i, j]:
                unknown_id[i, j] = k
                coords.append((i, j))
                k += 1

    rows = []
    cols = []
    data = []
    b = np.zeros(k, dtype=float)

    for eq, (i, j) in enumerate(coords):
        # z-direction harmonic face epsilons
        eps_zm = harmonic_mean(eps_rel[i, j], eps_rel[i, j - 1])
        eps_zp = harmonic_mean(eps_rel[i, j], eps_rel[i, j + 1])

        a_zm = eps_zm / dz**2
        a_zp = eps_zp / dz**2

        if i == 0:
            # Axis symmetry
            eps_rp = harmonic_mean(eps_rel[i, j], eps_rel[i + 1, j])
            a_rp = 2.0 * eps_rp / dr**2
            a_rm = 0.0
        else:
            ri = r[i]
            rph = ri + 0.5 * dr
            rmh = ri - 0.5 * dr

            eps_rp = harmonic_mean(eps_rel[i, j], eps_rel[i + 1, j])
            eps_rm = harmonic_mean(eps_rel[i, j], eps_rel[i - 1, j])

            a_rp = rph * eps_rp / (ri * dr**2)
            a_rm = rmh * eps_rm / (ri * dr**2)

        a_0 = -(a_rp + a_rm + a_zp + a_zm)

        rows.append(eq)
        cols.append(eq)
        data.append(a_0)

        def add_neighbor(ii, jj, coeff):
            if is_dir[ii, jj]:
                b[eq] -= coeff * V_bc[ii, jj]
            else:
                rows.append(eq)
                cols.append(unknown_id[ii, jj])
                data.append(coeff)

        if i == 0:
            add_neighbor(i + 1, j, a_rp)
        else:
            add_neighbor(i + 1, j, a_rp)
            add_neighbor(i - 1, j, a_rm)

        add_neighbor(i, j + 1, a_zp)
        add_neighbor(i, j - 1, a_zm)

    A = sparse.csr_matrix((data, (rows, cols)), shape=(k, k))
    return A, b, unknown_id, coords


def scatter_solution_to_grid(x, is_dir, V_bc, unknown_id):
    V = V_bc.copy()
    mask = ~is_dir
    V[mask] = x[unknown_id[mask]]
    return V


def solve_direct(A, b, is_dir, V_bc, unknown_id):
    t0 = time.perf_counter()
    x = spsolve(A, b)
    elapsed = time.perf_counter() - t0
    V = scatter_solution_to_grid(x, is_dir, V_bc, unknown_id)
    return V, elapsed


def solve_bicgstab(A, b, is_dir, V_bc, unknown_id, tol=1e-10, maxiter=5000):
    t0 = time.perf_counter()

    ilu = spilu(A.tocsc(), drop_tol=1e-6, fill_factor=10)
    M = LinearOperator(A.shape, ilu.solve)

    try:
        x, info = bicgstab(A, b, rtol=tol, atol=0.0, maxiter=maxiter, M=M)
    except TypeError:
        x, info = bicgstab(A, b, tol=tol, maxiter=maxiter, M=M)

    elapsed = time.perf_counter() - t0
    V = scatter_solution_to_grid(x, is_dir, V_bc, unknown_id)
    return V, elapsed, info


def compute_electric_field(V, dr, dz):
    dVdr, dVdz = np.gradient(V, dr, dz, edge_order=2)
    Er = -dVdr
    Ez = -dVdz
    Emag = np.sqrt(Er**2 + Ez**2)
    return Er, Ez, Emag
```


## Fowler–Nordheim post-processing

Emission is evaluated on CNT conductor faces that are directly exposed to vacuum.

Two important cautions apply:

1. The surface is represented on a structured grid, so the face-based CNT surface is still a staircase approximation.
2. The emitted current is exponentially sensitive to the local field, so **emission current converges more slowly** than lens-scale electrostatic observables.

That is why the notebook reports both:
- the **electrostatic convergence diagnostics** for Milestone 1, and
- the **emission outputs** for Milestone 2, without pretending that the current is already fully mesh-independent.


```python

def fowler_nordheim_current_density(F_local, phi_eV=4.8, beta_subgrid=1.0):
    '''
    Standard Fowler–Nordheim current density, optionally using an unresolved
    nanoscale enhancement factor beta_subgrid.
    '''
    A_FN = 1.541434e-6   # A eV V^-2
    B_FN = 6.830890e9    # V m^-1 eV^-3/2

    F_eff = beta_subgrid * np.asarray(F_local, dtype=float)
    J = np.zeros_like(F_eff, dtype=float)

    mask = F_eff > 0.0
    Fm = F_eff[mask]

    J[mask] = (A_FN / phi_eV) * Fm**2 * np.exp(-B_FN * phi_eV**1.5 / Fm)
    return J


def annulus_area_at_node(r, dr, i):
    r_in = max(0.0, r[i] - 0.5 * dr)
    r_out = r[i] + 0.5 * dr
    return math.pi * (r_out**2 - r_in**2)


def compute_cnt_surface_emission(V, r, z, conductor_id, material_id,
                                 cnt_id=1, vacuum_id=0,
                                 phi_eV=4.8, beta_subgrid=1.0):
    '''
    Surface-face emission on CNT conductor faces that touch vacuum.
    '''
    dr = r[1] - r[0]
    dz = z[1] - z[0]
    Nr, Nz = V.shape

    face_records = []

    for i in range(Nr):
        for j in range(Nz):
            if conductor_id[i, j] != cnt_id:
                continue

            Vc = V[i, j]

            neighbors = [
                (i,   j + 1, "zp"),
                (i+1, j,     "rp"),
                (i,   j - 1, "zm"),
                (i-1, j,     "rm"),
            ]

            for ii, jj, face in neighbors:
                if ii < 0 or ii >= Nr or jj < 0 or jj >= Nz:
                    continue

                if conductor_id[ii, jj] != 0:
                    continue

                if material_id[ii, jj] != vacuum_id:
                    continue

                Vv = V[ii, jj]

                if face in ("zp", "zm"):
                    F_local = abs(Vv - Vc) / dz
                    area = annulus_area_at_node(r, dr, i)
                    r_loc = r[i]
                    z_loc = 0.5 * (z[j] + z[jj])
                    normal_dir = "z"
                else:
                    r_face = 0.5 * (r[i] + r[ii])
                    F_local = abs(Vv - Vc) / dr
                    area = 2.0 * math.pi * r_face * dz
                    r_loc = r_face
                    z_loc = z[j]
                    normal_dir = "r"

                face_records.append((r_loc, z_loc, F_local, area, normal_dir))

    if len(face_records) == 0:
        return {
            "faces_df": pd.DataFrame(columns=["r", "z", "F_local", "area", "normal", "J_FN", "I_face"]),
            "I_total": 0.0,
            "F_local_max": 0.0,
            "J_max": 0.0,
        }

    faces_df = pd.DataFrame(face_records, columns=["r", "z", "F_local", "area", "normal"])
    faces_df["J_FN"] = fowler_nordheim_current_density(
        faces_df["F_local"].values,
        phi_eV=phi_eV,
        beta_subgrid=beta_subgrid,
    )
    faces_df["I_face"] = faces_df["J_FN"] * faces_df["area"]

    return {
        "faces_df": faces_df,
        "I_total": faces_df["I_face"].sum(),
        "F_local_max": faces_df["F_local"].max(),
        "J_max": faces_df["J_FN"].max(),
    }


def plot_cnt_emission_faces(emission):
    faces_df = emission["faces_df"]

    if len(faces_df) == 0:
        print("No CNT faces touching vacuum were found.")
        return

    positive = faces_df["J_FN"].values > 0
    if not np.any(positive):
        print("Emission current density is zero everywhere for the current parameters.")
        return

    fig, axs = plt.subplots(1, 2, figsize=(13, 4.5), constrained_layout=True)

    sc0 = axs[0].scatter(
        faces_df["z"] * 1e3,
        faces_df["r"] * 1e3,
        c=faces_df["F_local"],
        s=20,
        cmap="viridis",
    )
    fig.colorbar(sc0, ax=axs[0], label=r"Local surface field $F_n$ [V/m]")
    axs[0].set_xlabel("z [mm]")
    axs[0].set_ylabel("r [mm]")
    axs[0].set_title("CNT surface faces colored by local field")

    Jpos = faces_df.loc[positive, "J_FN"].values
    sc1 = axs[1].scatter(
        faces_df.loc[positive, "z"] * 1e3,
        faces_df.loc[positive, "r"] * 1e3,
        c=Jpos,
        s=20,
        cmap="magma",
        norm=LogNorm(vmin=max(np.percentile(Jpos, 5), 1e-30), vmax=max(np.percentile(Jpos, 99.5), 1.01 * np.percentile(Jpos, 5)))
    )
    fig.colorbar(sc1, ax=axs[1], label=r"$J_{\mathrm{FN}}$ [A/m$^2$]")
    axs[1].set_xlabel("z [mm]")
    axs[1].set_ylabel("r [mm]")
    axs[1].set_title("CNT emission map (log scale)")

    plt.show()
```

```python

def build_case(R_outer_case, z_extra_case, Nr_case, Nz_case, params, run_iterative=False):
    z_max_case = params["z3_end"] + z_extra_case

    r = np.linspace(0.0, R_outer_case, Nr_case)
    z = np.linspace(0.0, z_max_case, Nz_case)

    dr = r[1] - r[0]
    dz = z[1] - z[0]

    geom = build_geometry(r, z, params)
    A, b, unknown_id, coords = build_axisymmetric_variable_eps_matrix(
        r, z, geom["eps_rel"], geom["is_dir"], geom["V_bc"]
    )

    V_direct, t_direct = solve_direct(A, b, geom["is_dir"], geom["V_bc"], unknown_id)
    Er_direct, Ez_direct, E_direct = compute_electric_field(V_direct, dr, dz)

    emission = compute_cnt_surface_emission(
        V_direct, r, z,
        geom["conductor_id"], geom["material_id"],
        cnt_id=1, vacuum_id=0,
        phi_eV=params["work_function_eV"],
        beta_subgrid=params["beta_subgrid"],
    )

    j_apex = np.argmin(np.abs(z - params["h_cnt_total"]))
    apex_axis_field = abs(Ez_direct[0, min(j_apex + 1, len(z) - 1)])

    z_interest = z[z <= params["z_compare_end"]]
    V_axis = np.interp(z_interest, z, V_direct[0, :])
    Ez_axis = np.interp(z_interest, z, Ez_direct[0, :])

    observables = {
        "axisV_peak": V_axis.max(),
        "z_axisV_peak": z_interest[np.argmax(V_axis)],
        "axis_field_integral": np.trapezoid(Ez_axis, z_interest),
        "apex_axis_field": apex_axis_field,
        "I_FN_total": emission["I_total"],
        "CNT_surface_field_max": emission["F_local_max"],
    }

    out = {
        "r": r,
        "z": z,
        "dr": dr,
        "dz": dz,
        "geom": geom,
        "A": A,
        "b": b,
        "unknown_id": unknown_id,
        "V_direct": V_direct,
        "Er_direct": Er_direct,
        "Ez_direct": Ez_direct,
        "E_direct": E_direct,
        "t_direct": t_direct,
        "emission": emission,
        "observables": observables,
        "R_outer": R_outer_case,
        "z_extra": z_extra_case,
        "Nr": Nr_case,
        "Nz": Nz_case,
    }

    if run_iterative:
        V_bicg, t_bicg, info_bicg = solve_bicgstab(
            A, b, geom["is_dir"], geom["V_bc"], unknown_id,
            tol=bicg_tol, maxiter=bicg_maxiter
        )
        Er_bicg, Ez_bicg, E_bicg = compute_electric_field(V_bicg, dr, dz)
        out.update({
            "V_bicg": V_bicg,
            "Er_bicg": Er_bicg,
            "Ez_bicg": Ez_bicg,
            "E_bicg": E_bicg,
            "t_bicg": t_bicg,
            "bicg_info": info_bicg,
        })

    return out
```

```python

def plot_potential_and_field(case, title_prefix="Baseline"):
    r = case["r"]
    z = case["z"]
    V = case["V_direct"]
    E = case["E_direct"]
    conductor_mask = case["geom"]["conductor_id"] > 0

    fig, axs = plt.subplots(1, 2, figsize=(14, 4.8), constrained_layout=True)

    pcm0 = axs[0].pcolormesh(z * 1e3, r * 1e3, V, shading="auto", cmap="coolwarm")
    fig.colorbar(pcm0, ax=axs[0], label="V [V]")
    axs[0].contour(z * 1e3, r * 1e3, conductor_mask.astype(float), levels=[0.5], colors="k", linewidths=0.8)
    axs[0].contour(z * 1e3, r * 1e3, (case["geom"]["material_id"] == 1).astype(float), levels=[0.5], colors="cyan", linewidths=0.8)
    axs[0].set_title(f"{title_prefix}: potential")
    axs[0].set_xlabel("z [mm]")
    axs[0].set_ylabel("r [mm]")

    E_masked = np.ma.array(E, mask=conductor_mask)
    vals = E_masked.compressed()
    vmin = max(np.percentile(vals, 5), 1.0)
    vmax = max(np.percentile(vals, 99.5), 10.0 * vmin)

    pcm1 = axs[1].pcolormesh(
        z * 1e3, r * 1e3, E_masked,
        shading="auto",
        cmap="inferno",
        norm=LogNorm(vmin=vmin, vmax=vmax),
    )
    fig.colorbar(pcm1, ax=axs[1], label=r"$|\mathbf{E}|$ [V/m] (log scale)")
    axs[1].contour(z * 1e3, r * 1e3, conductor_mask.astype(float), levels=[0.5], colors="cyan", linewidths=0.7)
    axs[1].contour(z * 1e3, r * 1e3, (case["geom"]["material_id"] == 1).astype(float), levels=[0.5], colors="white", linewidths=0.7)
    axs[1].set_title(f"{title_prefix}: electric-field magnitude")
    axs[1].set_xlabel("z [mm]")
    axs[1].set_ylabel("r [mm]")

    plt.show()


def draw_electrode_spans(ax, params):
    ax.axvspan(params["z1_start"] * 1e3, params["z1_end"] * 1e3, color="gray", alpha=0.15)
    ax.axvspan(params["z2_start"] * 1e3, params["z2_end"] * 1e3, color="red",  alpha=0.10)
    ax.axvspan(params["z3_start"] * 1e3, params["z3_end"] * 1e3, color="gray", alpha=0.15)


def plot_axis_observables(case, params, title_prefix="Baseline"):
    z = case["z"]
    V_axis = case["V_direct"][0, :]
    Ez_axis = case["Ez_direct"][0, :]

    fig, axs = plt.subplots(2, 1, figsize=(8.2, 7.0), sharex=True, constrained_layout=True)

    axs[0].plot(z * 1e3, V_axis, lw=2)
    draw_electrode_spans(axs[0], params)
    axs[0].set_ylabel(r"$V(0,z)$ [V]")
    axs[0].set_title(f"{title_prefix}: axis potential")
    axs[0].grid(True, alpha=0.3)

    axs[1].plot(z * 1e3, Ez_axis, lw=2)
    draw_electrode_spans(axs[1], params)
    axs[1].axhline(0.0, color="k", lw=1, alpha=0.5)
    axs[1].set_xlabel("z [mm]")
    axs[1].set_ylabel(r"$E_z(0,z)$ [V/m]")
    axs[1].set_title(f"{title_prefix}: axis electric field")
    axs[1].grid(True, alpha=0.3)

    plt.show()
```


## Milestone 1 diagnostics: boundary sensitivity

The boundary study keeps the **local mesh spacing fixed** while expanding the outer radius and/or the top truncation distance. This is important: otherwise a domain-change study becomes contaminated by a simultaneous mesh-change study.

For each case, the notebook compares to the baseline on:

- relative $L^2$ difference of the axis potential in the lens region,
- relative difference of the integrated axis field,
- relative difference of the peak axis potential,
- relative difference of the apex-axis field.

Those are used as the boundary-sensitivity pass/fail criteria.


```python

def rel_l2(a, b):
    return np.linalg.norm(a - b) / np.linalg.norm(a)


def axis_window(case, z_end):
    z_sel = case["z"][case["z"] <= z_end]
    V_axis = np.interp(z_sel, case["z"], case["V_direct"][0, :])
    Ez_axis = np.interp(z_sel, case["z"], case["Ez_direct"][0, :])
    return z_sel, V_axis, Ez_axis


def baseline_spacing(case):
    return case["dr"], case["dz"]


def nodes_for_same_spacing(R_case, z_extra_case, params, dr_target, dz_target):
    z_max_case = params["z3_end"] + z_extra_case
    Nr_case = int(round(R_case / dr_target)) + 1
    Nz_case = int(round(z_max_case / dz_target)) + 1
    return Nr_case, Nz_case


def boundary_study(params, boundary_cases, baseline_case):
    dr0, dz0 = baseline_spacing(baseline_case)
    z_end = params["z_compare_end"]

    z_ref, V_ref, Ez_ref = axis_window(baseline_case, z_end)
    field_int_ref = np.trapezoid(Ez_ref, z_ref)
    peakV_ref = V_ref.max()
    apexF_ref = baseline_case["observables"]["apex_axis_field"]

    results = {}
    rows = []

    for name, (R_case, z_extra_case) in boundary_cases.items():
        if name == "baseline":
            case = baseline_case
        else:
            Nr_case, Nz_case = nodes_for_same_spacing(R_case, z_extra_case, params, dr0, dz0)
            case = build_case(
                R_outer_case=R_case,
                z_extra_case=z_extra_case,
                Nr_case=Nr_case,
                Nz_case=Nz_case,
                params=params,
                run_iterative=False,
            )

        z_cmp, V_cmp, Ez_cmp = axis_window(case, z_end)
        V_interp = np.interp(z_ref, z_cmp, V_cmp)
        Ez_interp = np.interp(z_ref, z_cmp, Ez_cmp)

        field_int_cmp = np.trapezoid(Ez_interp, z_ref)
        peakV_cmp = V_interp.max()
        apexF_cmp = case["observables"]["apex_axis_field"]

        metrics = {
            "R_outer_mm": case["R_outer"] * 1e3,
            "z_extra_mm": case["z_extra"] * 1e3,
            "Nr": case["Nr"],
            "Nz": case["Nz"],
            "t_direct_s": case["t_direct"],
            "relL2_axisV": rel_l2(V_ref, V_interp),
            "reldiff_axis_field_integral": abs(field_int_cmp - field_int_ref) / abs(field_int_ref),
            "reldiff_peak_axisV": abs(peakV_cmp - peakV_ref) / abs(peakV_ref),
            "reldiff_apex_axis_field": abs(apexF_cmp - apexF_ref) / abs(apexF_ref),
        }

        results[name] = {"case": case, "metrics": metrics}
        rows.append(pd.Series(metrics, name=name))

    report = pd.DataFrame(rows)
    return results, report


boundary_results, boundary_report = boundary_study(params, boundary_cases, baseline)

boundary_report
```

```python

boundary_report["pass_axisV"] = boundary_report["relL2_axisV"] <= tol_boundary_axisV
boundary_report["pass_field_int"] = boundary_report["reldiff_axis_field_integral"] <= tol_boundary_field_int
boundary_report["pass_peakV"] = boundary_report["reldiff_peak_axisV"] <= tol_boundary_peakV
boundary_report["pass_apexF"] = boundary_report["reldiff_apex_axis_field"] <= tol_boundary_apexF

display(boundary_report)

z_end = params["z_compare_end"]
z_ref, V_ref, Ez_ref = axis_window(baseline, z_end)

plt.figure(figsize=(8.0, 4.6))
for name, entry in boundary_results.items():
    z_cmp, V_cmp, _ = axis_window(entry["case"], z_end)
    plt.plot(z_ref * 1e3, np.interp(z_ref, z_cmp, V_cmp), label=name)
draw_electrode_spans(plt.gca(), params)
plt.xlabel("z [mm]")
plt.ylabel(r"$V(0,z)$ [V]")
plt.title("Boundary study: axis potential")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```


## Milestone 1 diagnostics: mesh refinement

The mesh study holds the physical domain fixed and refines the grid.

Again, the pass/fail criteria are based on **lens-scale observables** rather than the absolute strongest corner field:

- relative $L^2$ change in axis potential,
- relative change in the integrated axis field,
- relative change in the peak axis potential.

The apex field is reported too, but it is interpreted more cautiously because local surface-field convergence is slower.


```python

def mesh_refinement_study(params, mesh_grids, R_outer, z_extra):
    cases = {}
    rows = []

    for name, (Nr_case, Nz_case) in mesh_grids.items():
        case = build_case(
            R_outer_case=R_outer,
            z_extra_case=z_extra,
            Nr_case=Nr_case,
            Nz_case=Nz_case,
            params=params,
            run_iterative=False,
        )
        cases[name] = case

        row = {
            "Nr": Nr_case,
            "Nz": Nz_case,
            "t_direct_s": case["t_direct"],
            "axisV_peak_V": case["observables"]["axisV_peak"],
            "z_axisV_peak_mm": case["observables"]["z_axisV_peak"] * 1e3,
            "axis_field_integral_V": case["observables"]["axis_field_integral"],
            "apex_axis_field_Vpm": case["observables"]["apex_axis_field"],
            "I_FN_total_A": case["observables"]["I_FN_total"],
            "CNT_surface_field_max_Vpm": case["observables"]["CNT_surface_field_max"],
        }
        rows.append(pd.Series(row, name=name))

    report = pd.DataFrame(rows)
    return cases, report


mesh_cases, mesh_report = mesh_refinement_study(params, mesh_grids, R_outer, z_extra)
mesh_report
```

```python

# Compare all grids against the xfine reference for lens-scale observables
reference_name = "xfine"
reference_case = mesh_cases[reference_name]

z_end = params["z_compare_end"]
z_ref, V_ref, Ez_ref = axis_window(reference_case, z_end)
field_int_ref = np.trapezoid(Ez_ref, z_ref)
peakV_ref = V_ref.max()
apexF_ref = reference_case["observables"]["apex_axis_field"]

comparison_rows = []

for name, case in mesh_cases.items():
    z_cmp, V_cmp, Ez_cmp = axis_window(case, z_end)
    V_interp = np.interp(z_ref, z_cmp, V_cmp)
    Ez_interp = np.interp(z_ref, z_cmp, Ez_cmp)

    field_int_cmp = np.trapezoid(Ez_interp, z_ref)
    peakV_cmp = V_interp.max()
    apexF_cmp = case["observables"]["apex_axis_field"]

    comparison_rows.append(pd.Series({
        "relL2_axisV_vs_xfine": rel_l2(V_ref, V_interp),
        "reldiff_axis_field_integral_vs_xfine": abs(field_int_cmp - field_int_ref) / abs(field_int_ref),
        "reldiff_peak_axisV_vs_xfine": abs(peakV_cmp - peakV_ref) / abs(peakV_ref),
        "reldiff_apex_axis_field_vs_xfine": abs(apexF_cmp - apexF_ref) / abs(apexF_ref),
    }, name=name))

mesh_compare = pd.DataFrame(comparison_rows)

mesh_compare["pass_axisV"] = mesh_compare["relL2_axisV_vs_xfine"] <= tol_mesh_axisV
mesh_compare["pass_field_int"] = mesh_compare["reldiff_axis_field_integral_vs_xfine"] <= tol_mesh_field_int
mesh_compare["pass_peakV"] = mesh_compare["reldiff_peak_axisV_vs_xfine"] <= tol_mesh_peakV
mesh_compare["pass_apexF"] = mesh_compare["reldiff_apex_axis_field_vs_xfine"] <= tol_mesh_apexF

display(mesh_compare)

plt.figure(figsize=(8.0, 4.6))
for name, case in mesh_cases.items():
    z_cmp, V_cmp, _ = axis_window(case, z_end)
    plt.plot(z_ref * 1e3, np.interp(z_ref, z_cmp, V_cmp), label=name)
draw_electrode_spans(plt.gca(), params)
plt.xlabel("z [mm]")
plt.ylabel(r"$V(0,z)$ [V]")
plt.title("Mesh refinement: axis potential")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```


## Interpretation of what this notebook accomplishes

### Milestone 1: Defensible electrostatics
Implemented here:
- explicit dielectric through $\nabla\cdot(\epsilon\nabla V)=0$,
- finite-thickness and rounded ring electrodes,
- effective protruding CNT geometry,
- domain-truncation sensitivity study,
- mesh-refinement study,
- convergence metrics on lens-scale electrostatic observables.

### Milestone 2: Emission coupling
Implemented here:
- local CNT surface field extraction,
- Fowler–Nordheim current-density evaluation,
- axisymmetric surface-current integration,
- optional sub-grid enhancement factor for unresolved CNT nanoscale sharpening.

### What still remains for the next notebook
The next notebook should build on the saved baseline arrays and add:

- electron trajectory integration in the computed $(r,z)$ field,
- focusing / spot-size observables,
- optional multi-particle emission sampling from the Fowler–Nordheim surface map.
