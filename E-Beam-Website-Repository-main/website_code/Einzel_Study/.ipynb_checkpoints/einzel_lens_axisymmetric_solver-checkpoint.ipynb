
# Axisymmetric electrostatic solver for a CNT-field-emission Einzel lens

This notebook builds a **2D axisymmetric** electrostatic model for the geometry in your sketch and computes the potential
\(V(r,z)\) and electric field \(\mathbf{E}(r,z)=-\nabla V\) on a numerical grid.

## Physical model and interpretation of the sketch

The sketch is most naturally interpreted as an **axisymmetric cylindrical** problem, not a spherically symmetric one:

- axial coordinate: \(z\), measured upward from the CNT/cathode plane
- radial coordinate: \(r\in[0,R]\), where \(R\) is the aperture / lens radius
- unknown: electrostatic potential \(V(r,z)\)

In vacuum, with no free charge in the computational domain, the potential satisfies the **axisymmetric Laplace equation**
\[
\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial V}{\partial r}\right) + \frac{\partial^2 V}{\partial z^2}=0.
\]

We solve this on a rectangular computational domain in \((r,z)\):

- \(0 \le r \le R\)
- \(0 \le z \le z_\max\)

with boundary conditions chosen to approximate the CNT emitter, the three-electrode einzel lens, and a far-field region above the final electrode.

## Geometry assumption used here

Because the hand sketch labels only the key distances and voltages, this notebook uses a standard reduced model:

- the **bottom plane** at \(z=0\) is the CNT/cathode plane at \(-V_\mathrm{CNT}\)
- the **outer cylindrical wall** at \(r=R\) carries three thin electrode segments:
  - first electrode at 0 V
  - middle electrode at \(+V_\mathrm{mid}\)
  - final electrode at 0 V
- the gaps between those electrode segments are the labeled distances \(d_1,d_2,d_3\)
- above the final grounded electrode, the domain is extended to \(z_\max\) so the field can decay

This is a common axisymmetric surrogate for an aperture lens. It gives a full 2D array \(V[r_i,z_j]\) and allows direct field extraction.

## Numerical methods implemented

The notebook solves the same discretized problem with three methods:

1. **Sparse direct solve (`spsolve`)**  
   Most accurate / robust reference on modest grids. Good benchmark.

2. **Preconditioned Krylov solve (`cg`)**  
   Uses the same matrix but solves iteratively. Faster and more memory-efficient on larger grids.

3. **SOR (Successive Over-Relaxation)**  
   Classic finite-difference iterative method. Included as a transparent baseline and for method comparison.

We then:

- store the resulting 2D potential arrays in memory
- compute electric fields from the potential
- plot heat maps
- compare runtimes and solution differences

## Notes on “state of the art”

For production electrostatics, the strongest general-purpose approaches are typically:

- sparse direct factorizations on moderate meshes
- multigrid / algebraic multigrid preconditioned Krylov methods on large meshes
- FEM packages such as PETSc/FEniCSx, COMSOL, Elmer, MFEM, etc.

This notebook stays fully self-contained in standard scientific Python and uses methods that are robust and reproducible in a normal Jupyter environment. The finite-volume-style axisymmetric discretization used below is much better than a naive Cartesian approximation.


```python

import time
from dataclasses import dataclass

import numpy as np
import matplotlib.pyplot as plt

from scipy import sparse
from scipy.sparse.linalg import spsolve, cg, spilu, LinearOperator
```


## 1. User parameters

Edit the parameters below to match your physical geometry.

Units are SI unless otherwise noted.


```python

# -----------------------------
# Geometry and voltage settings
# -----------------------------
R = 0.5e-3          # lens radius / aperture radius [m]
d1 = 0.4e-3         # cathode-to-electrode-1 gap [m]
d2 = 0.5e-3         # electrode-1 to electrode-2 gap [m]
d3 = 0.5e-3         # electrode-2 to electrode-3 gap [m]
t_e = 0.1e-3        # electrode thickness [m]
z_extra = 1.0e-3    # extra distance above final electrode [m]

V_cnt = 800.0       # cathode magnitude [V] => cathode potential = -V_cnt
V_mid = 500.0       # middle lens electrode [V]
V_ground = 0.0      # grounded electrodes [V]

# -----------------------------
# Numerical grid
# -----------------------------
Nr = 121            # number of radial nodes (including r=0 and r=R)
Nz = 321            # number of axial nodes (including z=0 and z=z_max)

# -----------------------------
# Solver settings
# -----------------------------
cg_tol = 1e-10
cg_maxiter = 5000

sor_omega = 1.85
sor_tol = 1e-8
sor_maxiter = 25000
report_every = 200
```

```python

z1_start = d1
z1_end   = z1_start + t_e

z2_start = z1_end + d2
z2_end   = z2_start + t_e

z3_start = z2_end + d3
z3_end   = z3_start + t_e

z_max = z3_end + z_extra

print(f"z_max = {z_max*1e3:.3f} mm")
print("Electrode segments on outer wall:")
print(f"  Electrode 1: {z1_start*1e3:.3f} to {z1_end*1e3:.3f} mm at 0 V")
print(f"  Electrode 2: {z2_start*1e3:.3f} to {z2_end*1e3:.3f} mm at +V_mid")
print(f"  Electrode 3: {z3_start*1e3:.3f} to {z3_end*1e3:.3f} mm at 0 V")
```


## 2. Grid and boundary-condition masks

We store the solution on a 2D grid shaped `(Nr, Nz)` with axis ordering `(r, z)`.

Boundary conditions used here:

- `z = 0`: cathode plane, \(V=-V_\mathrm{CNT}\)
- `r = 0`: symmetry axis, \(\partial V / \partial r = 0\)
- `r = R`: piecewise Dirichlet wall, modeling the einzel-lens electrodes and grounded far-field wall
- `z = z_max`: grounded far-field cap (a practical truncation of the infinite domain)

This gives a fully closed problem suitable for sparse linear algebra.


```python

r = np.linspace(0.0, R, Nr)
z = np.linspace(0.0, z_max, Nz)

dr = r[1] - r[0]
dz = z[1] - z[0]

RR, ZZ = np.meshgrid(r, z, indexing="ij")

def make_dirichlet_bc_mask(r, z, V_cnt, V_mid, V_ground):
    Nr, Nz = len(r), len(z)
    is_dir = np.zeros((Nr, Nz), dtype=bool)
    V_bc = np.zeros((Nr, Nz), dtype=float)

    # Bottom cathode plane
    is_dir[:, 0] = True
    V_bc[:, 0] = -V_cnt

    # Top far-field cap
    is_dir[:, -1] = True
    V_bc[:, -1] = V_ground

    # Outer radial wall: grounded except for the middle electrode segment
    is_dir[-1, :] = True
    V_bc[-1, :] = V_ground

    z_arr = z
    e1 = (z_arr >= z1_start) & (z_arr <= z1_end)
    e2 = (z_arr >= z2_start) & (z_arr <= z2_end)
    e3 = (z_arr >= z3_start) & (z_arr <= z3_end)

    V_bc[-1, e1] = V_ground
    V_bc[-1, e2] = V_mid
    V_bc[-1, e3] = V_ground

    # Axis r=0 is NOT Dirichlet; it uses symmetry (Neumann)
    return is_dir, V_bc

is_dir, V_bc = make_dirichlet_bc_mask(r, z, V_cnt, V_mid, V_ground)

plt.figure(figsize=(7,4))
plt.pcolormesh(z*1e3, r*1e3, np.where(is_dir, V_bc, np.nan), shading="auto")
plt.colorbar(label="Dirichlet boundary value [V]")
plt.xlabel("z [mm]")
plt.ylabel("r [mm]")
plt.title("Boundary-condition map (NaN = interior / symmetry axis)")
plt.tight_layout()
plt.show()
```


## 3. Axisymmetric finite-volume / finite-difference discretization

For interior nodes we discretize

\[
\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial V}{\partial r}\right)+\frac{\partial^2V}{\partial z^2}=0
\]

in a flux-conservative way. At a node \((r_i,z_j)\) with \(r_i>0\),

\[
\frac{1}{r_i\,\Delta r^2}
\left[
r_{i+1/2}(V_{i+1,j}-V_{i,j}) - r_{i-1/2}(V_{i,j}-V_{i-1,j})
\right]
+
\frac{V_{i,j+1}-2V_{i,j}+V_{i,j-1}}{\Delta z^2}
=0.
\]

At the symmetry axis \(r=0\), we impose \(\partial V/\partial r=0\), which yields the standard axis formula

\[
\left.\frac{1}{r}\frac{\partial}{\partial r}\left(r\frac{\partial V}{\partial r}\right)\right|_{r=0}
\approx
\frac{2(V_{1,j}-V_{0,j})}{\Delta r^2}.
\]

This preserves the correct axisymmetric structure and is better than pretending the problem is Cartesian.


```python

def build_axisymmetric_matrix(r, z, is_dir, V_bc):
    '''
    Build A x = b for the axisymmetric Laplace equation on a structured grid.

    Unknown vector x contains only non-Dirichlet nodes.
    '''
    Nr, Nz = len(r), len(z)
    dr = r[1] - r[0]
    dz = z[1] - z[0]

    unknown_id = -np.ones((Nr, Nz), dtype=int)
    coords = []
    k = 0
    for i in range(Nr):
        for j in range(Nz):
            if not is_dir[i, j]:
                unknown_id[i, j] = k
                coords.append((i, j))
                k += 1

    rows = []
    cols = []
    data = []
    b = np.zeros(k, dtype=float)

    inv_dz2 = 1.0 / dz**2
    inv_dr2 = 1.0 / dr**2

    def add_to_equation(eq, i2, j2, coeff):
        if is_dir[i2, j2]:
            b[eq] -= coeff * V_bc[i2, j2]
        else:
            rows.append(eq)
            cols.append(unknown_id[i2, j2])
            data.append(coeff)

    for eq, (i, j) in enumerate(coords):
        # z-direction coefficients
        c_zm = inv_dz2
        c_zp = inv_dz2

        if i == 0:
            # symmetry axis
            c_rp = 2.0 * inv_dr2
            c_rm = 0.0
            c_0 = -(2.0 * inv_dr2 + 2.0 * inv_dz2)

            rows.append(eq); cols.append(eq); data.append(c_0)

            add_to_equation(eq, i+1, j, c_rp)
            add_to_equation(eq, i, j-1, c_zm)
            add_to_equation(eq, i, j+1, c_zp)

        else:
            ri = r[i]
            rph = ri + 0.5 * dr
            rmh = ri - 0.5 * dr

            c_rp = rph / (ri * dr**2)
            c_rm = rmh / (ri * dr**2)
            c_0 = -(c_rp + c_rm + 2.0 * inv_dz2)

            rows.append(eq); cols.append(eq); data.append(c_0)

            add_to_equation(eq, i+1, j, c_rp)
            add_to_equation(eq, i-1, j, c_rm)
            add_to_equation(eq, i, j-1, c_zm)
            add_to_equation(eq, i, j+1, c_zp)

    A = sparse.csr_matrix((data, (rows, cols)), shape=(k, k))
    return A, b, unknown_id, coords

A, b, unknown_id, coords = build_axisymmetric_matrix(r, z, is_dir, V_bc)
print("Matrix shape:", A.shape)
print("nnz:", A.nnz)
```


## 4. Solver implementations

We now solve the same linear system with three methods.

### 4.1 Sparse direct solve
This is the most reliable reference result on a grid of this size.

### 4.2 Preconditioned conjugate gradient
The discretized axisymmetric Laplace operator is symmetric positive definite after eliminating Dirichlet nodes.
A simple ILU-based preconditioner is used here through `spilu`.

### 4.3 Successive over-relaxation (SOR)
This is slower but useful pedagogically and for cross-checking.


```python

def scatter_solution_to_grid(x, is_dir, V_bc, unknown_id):
    V = V_bc.copy()
    mask = ~is_dir
    V[mask] = x[unknown_id[mask]]
    return V

def solve_direct(A, b, is_dir, V_bc, unknown_id):
    t0 = time.perf_counter()
    x = spsolve(A, b)
    elapsed = time.perf_counter() - t0
    V = scatter_solution_to_grid(x, is_dir, V_bc, unknown_id)
    return V, elapsed

def solve_cg(A, b, is_dir, V_bc, unknown_id, tol=1e-10, maxiter=5000):
    # ILU preconditioner
    t0 = time.perf_counter()
    ilu = spilu(A.tocsc(), drop_tol=1e-5, fill_factor=10)
    M = LinearOperator(A.shape, ilu.solve)

    x, info = cg(A, b, rtol=tol, atol=0.0, maxiter=maxiter, M=M)
    elapsed = time.perf_counter() - t0

    if info != 0:
        print(f"CG warning: info={info} (0 means successful convergence)")
    V = scatter_solution_to_grid(x, is_dir, V_bc, unknown_id)
    return V, elapsed, info
```

```python

def solve_sor(r, z, is_dir, V_bc, omega=1.85, tol=1e-8, maxiter=25000, report_every=200):
    Nr, Nz = len(r), len(z)
    dr = r[1] - r[0]
    dz = z[1] - z[0]
    inv_dz2 = 1.0 / dz**2

    V = V_bc.copy()

    # Initial guess in the interior: linear interpolation from bottom to top
    for j in range(Nz):
        frac = z[j] / z[-1]
        guess = (1 - frac) * (-V_cnt) + frac * 0.0
        for i in range(Nr):
            if not is_dir[i, j]:
                V[i, j] = guess

    history = []
    t0 = time.perf_counter()

    for it in range(1, maxiter + 1):
        max_update = 0.0

        for i in range(Nr):
            for j in range(1, Nz - 1):
                if is_dir[i, j]:
                    continue

                if i == 0:
                    # axis node
                    a_p = 2.0 / dr**2
                    a_z = 1.0 / dz**2
                    denom = 2.0 / dr**2 + 2.0 / dz**2
                    V_new = (a_p * V[i+1, j] + a_z * V[i, j-1] + a_z * V[i, j+1]) / denom
                else:
                    ri = r[i]
                    rph = ri + 0.5 * dr
                    rmh = ri - 0.5 * dr
                    a_p = rph / (ri * dr**2)
                    a_m = rmh / (ri * dr**2)
                    denom = a_p + a_m + 2.0 / dz**2
                    V_new = (
                        a_p * V[i+1, j] +
                        a_m * V[i-1, j] +
                        inv_dz2 * V[i, j-1] +
                        inv_dz2 * V[i, j+1]
                    ) / denom

                delta = V_new - V[i, j]
                V[i, j] += omega * delta
                max_update = max(max_update, abs(omega * delta))

        history.append(max_update)
        if (it % report_every == 0) or (it == 1):
            print(f"SOR iter {it:6d}: max update = {max_update:.3e} V")

        if max_update < tol:
            elapsed = time.perf_counter() - t0
            print(f"SOR converged in {it} iterations")
            return V, elapsed, it, np.array(history)

    elapsed = time.perf_counter() - t0
    print("SOR reached maxiter without satisfying tolerance")
    return V, elapsed, maxiter, np.array(history)
```


## 5. Run all solvers


```python

V_direct, t_direct = solve_direct(A, b, is_dir, V_bc, unknown_id)
V_cg, t_cg, cg_info = solve_cg(A, b, is_dir, V_bc, unknown_id, tol=cg_tol, maxiter=cg_maxiter)
V_sor, t_sor, it_sor, sor_history = solve_sor(
    r, z, is_dir, V_bc,
    omega=sor_omega, tol=sor_tol, maxiter=sor_maxiter, report_every=report_every
)

print(f"Direct solve time: {t_direct:.4f} s")
print(f"CG solve time:     {t_cg:.4f} s")
print(f"SOR solve time:    {t_sor:.4f} s")
```


## 6. Electric field extraction

Once we have \(V(r,z)\), we compute the electric field from

\[
E_r = -\frac{\partial V}{\partial r}, \qquad E_z = -\frac{\partial V}{\partial z}.
\]

We use second-order finite differences through `numpy.gradient`, which is a sensible post-processing choice on a structured grid.


```python

def compute_electric_field(V, dr, dz):
    dVdr, dVdz = np.gradient(V, dr, dz, edge_order=2)
    Er = -dVdr
    Ez = -dVdz
    Emag = np.sqrt(Er**2 + Ez**2)
    return Er, Ez, Emag

Er_direct, Ez_direct, E_direct = compute_electric_field(V_direct, dr, dz)
Er_cg, Ez_cg, E_cg = compute_electric_field(V_cg, dr, dz)
Er_sor, Ez_sor, E_sor = compute_electric_field(V_sor, dr, dz)
```


## 7. Store the 2D arrays in memory and optionally on disk

You can also save them to an `.npz` file for later use.


```python

results = {
    "r": r,
    "z": z,
    "V_direct": V_direct,
    "V_cg": V_cg,
    "V_sor": V_sor,
    "Er_direct": Er_direct,
    "Ez_direct": Ez_direct,
    "E_direct": E_direct,
    "Er_cg": Er_cg,
    "Ez_cg": Ez_cg,
    "E_cg": E_cg,
    "Er_sor": Er_sor,
    "Ez_sor": Ez_sor,
    "E_sor": E_sor,
}

save_arrays = True
if save_arrays:
    np.savez(
        "einzel_lens_fields.npz",
        **results,
    )
    print("Saved arrays to einzel_lens_fields.npz")
```


## 9. Method comparison

We compare:

- runtime
- relative L2 differences in potential
- relative L2 differences in electric-field magnitude

The direct solve is used as the reference.


```python

def rel_l2(A, B):
    return np.linalg.norm(A - B) / np.linalg.norm(A)

summary = {
    "direct_time_s": t_direct,
    "cg_time_s": t_cg,
    "sor_time_s": t_sor,
    "cg_vs_direct_relL2_V": rel_l2(V_direct, V_cg),
    "sor_vs_direct_relL2_V": rel_l2(V_direct, V_sor),
    "cg_vs_direct_relL2_E": rel_l2(E_direct, E_cg),
    "sor_vs_direct_relL2_E": rel_l2(E_direct, E_sor),
}

summary
```

```python

labels = ["Direct", "CG", "SOR"]
times = [t_direct, t_cg, t_sor]

plt.figure(figsize=(6,4))
plt.bar(labels, times)
plt.ylabel("Runtime [s]")
plt.title("Solver runtime comparison")
plt.tight_layout()
plt.show()

print("Relative differences w.r.t. direct solver")
print(f"CG  : rel-L2(V) = {summary['cg_vs_direct_relL2_V']:.3e}, rel-L2(E) = {summary['cg_vs_direct_relL2_E']:.3e}")
print(f"SOR : rel-L2(V) = {summary['sor_vs_direct_relL2_V']:.3e}, rel-L2(E) = {summary['sor_vs_direct_relL2_E']:.3e}")
```


## 10. Practical interpretation and next steps

### What this notebook already gives you
- a full **2D potential array** `V_direct`, `V_cg`, `V_sor`
- a full **2D electric-field array** `Er`, `Ez`, `|E|`
- heat maps and axis profiles
- solver-comparison diagnostics

### Important modeling caveat
This is a **reduced axisymmetric wall-electrode model** of the sketch. It is very useful for rapid computation and method comparison, but it is still an approximation.

### Stronger next upgrades
If you want to push this toward a more research-grade electrostatic model, the natural next improvements are:

1. **Explicit finite-thickness electrodes and dielectric regions**  
   Include the SiO2 region and solve
   \[
   \nabla\cdot(\epsilon\nabla V)=0
   \]
   with spatially varying permittivity.

2. **Local CNT-tip refinement**
   If the field enhancement near the CNT forest matters, refine strongly near the emitter or include an effective-emitter geometry.

3. **Adaptive mesh or FEM**
   For sharper edges and more geometric realism, move to FEM or an adaptive multigrid code.

4. **Particle tracing**
   Use the computed \(\mathbf{E}(r,z)\) as input for electron trajectory integration.

If you want, the next step can be a second notebook that:
- adds a dielectric region for SiO2,
- includes more realistic electrode geometry,
- or traces electron trajectories through the computed field.
